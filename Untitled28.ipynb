Data cleaning and preprocessing


```python
df_cleaned=df.copy()
```

```python
df_cleaned.drop_duplicates(inplace=True)
```

```python
df_cleaned.shape
```

```python
df_cleaned['sex']=df_cleaned['sex'].map({'male':0,'female':1})
```

```python
df_cleaned['smoker'].value_counts()
```

```python
df_cleaned['smoker']=df_cleaned['smoker'].map({'no':0,'yes':1})
```

```python
df_cleaned=pd.get_dummies(df_cleaned,columns=['region'],drop_first=True)
```

```python
df_cleaned
```

```python
df_cleaned.astype(int)
```

Feature Engineering and Extraction

```python
df_cleaned['bmi_cateogary']=pd.cut(
    df_cleaned['bmi'],
    bins=[0,18.5,24.9,29.9,np.inf],
    labels=['underweight','healthy','overweight','obesity']
)
```

```python
df_cleaned.head()
```

```python
df_cleaned['age_cateogary']=pd.cut(df_cleaned['age'],bins=[0,13,18,60,np.inf],labels=['child','teen','adult','senior'])
```

```python
df_cleaned
```

```python
df_cleaned=pd.get_dummies(df_cleaned,columns=['bmi_cateogary'],drop_first=True)
```

```python
df_cleaned=pd.get_dummies(df_cleaned,columns=['age_cateogary'],drop_first=True)
```

```python
df_cleaned=df_cleaned.astype(int)
```

```python
df_cleaned
```

```python
cols=['age','bmi','children']
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
df_cleaned[cols]=sc.fit_transform(df_cleaned[cols])
```
